```python
import pandas as pd
import os
```

```python
fake_data = pd.read_csv('./datasets/politifact_fake.csv')
real_data = pd.read_csv('./datasets/politifact_real.csv')

fake_data['labels'] = 1
real_data['labels'] = 0

data = pd.concat([real_data, fake_data], ignore_index=True)
```

```python
data.info()
```

```python
data.sample(10)
```

```python
IMAGE_ROOT_DIR = 'cleaned_politifact_imgs/'
images = []
for line in data.values:
    images_path = os.path.join(IMAGE_ROOT_DIR, str(line[0]))

    if os.path.exists(images_path):
        images.append(os.listdir(images_path))
    else:
        images.append(None)
print(images)
```

```python
data['images'] = images
```

```python
data.sample(10)
```

```python
data.dropna(subset=['images'], inplace=True)
```

```python
data.to_csv('datasets/politifact_cln_images.csv', index=False)
```
